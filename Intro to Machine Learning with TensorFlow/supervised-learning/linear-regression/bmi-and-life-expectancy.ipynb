```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%config InlineBackend.close_figures = False
```

```python
bmi_life_data = pd.read_csv('../../../data/bmi_and_life_expectancy.csv')
bmi_life_data
```

Extract the labeled data:

```python
X = bmi_life_data[['BMI']]
Y = bmi_life_data[['Life expectancy']]
```

```python
plt.scatter(X, Y)
plt.show()
```

Fit a linear regression:

```python
bmi_life_model = LinearRegression()
bmi_life_model.fit(X, Y)
```

```python
m = bmi_life_model.coef_[0][0]
b = bmi_life_model.intercept_[0]

"y = {:.2f}x + {:.2f}".format(m, b)
```

```python
X_min = X.min()
X_max = X.max()
plt.plot([X_min, X_max],[m * X_min + b, m * X_max + b], 'g')
plt.show()
```

Use fitted model to predict a value:

```python
laos_bmi = 21.07931
laos_life_exp = bmi_life_model.predict([ [laos_bmi] ])
laos_life_exp
```

```python
plt.plot(laos_bmi, laos_life_exp[0][0], 'r*', markersize=25)
plt.show()
```
