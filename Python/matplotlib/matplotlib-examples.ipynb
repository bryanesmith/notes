# Matplotlib

## Scatterplot

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

np.random.seed(42)
```

```python
a = np.floor(10 * np.random.random((100,2)))
a[0:5,] # first five rows
```

```python
X = a[:,0]
Y = a[:,1]
X_min, X_max = X.min(), X.max()
Y_min, Y_max = Y.min(), Y.max()

plt.scatter(X, Y, zorder = 3)
plt.plot([X_min, X_max],[.5 * Y_min + .5 * Y_max, Y_max])
plt.plot([X_min, .5 * X_min + .5 * X_max], [Y_max, Y_min])
plt.show()
```
